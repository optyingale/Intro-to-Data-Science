```python
import pandas as pd
df_pumas = pd.read_excel('pumas.xlsx', header = 0, index_col='TORNEO')
df_p = df_pumas[48:81]
del df_p['EFE (%)']
df_p.head()
```

```python
df_america = pd.read_excel('america.xlsx', header = 0, index_col='TORNEO')
df_a = df_america[88:121]
del df_a['EFE (%)']
df_a.head()
```

```python
df_cruz_azul = pd.read_excel('cruz_azul.xlsx', header = 0, index_col='TORNEO')
df_ca = df_cruz_azul[47:80]
df_ca.head()
```

```python
%matplotlib notebook
import matplotlib as mpl
mpl.get_backend()
```

```python
Percentage_Win_ca = df_ca.PG / df_ca.PJ
Percentage_Win_a = df_a.PG / df_a.PJ
Percentage_Win_p = df_p.PG / df_p.PJ
```

```python
import matplotlib
matplotlib.style.use('ggplot')
import matplotlib.pyplot as plt
plt.figure()
plt.plot(Percentage_Win_a,'gold',Percentage_Win_ca,'dodgerblue',Percentage_Win_p,'goldenrod')
plt.title("CDMX soccer teams")
plt.xlabel('Tournaments')
plt.ylabel('%WIN = Won Matches/Played Matches')
plt.legend(['América', 'Cruz Azul', 'Pumas'])

x = plt.gca().xaxis

# rotate the tick labels for the x axis
for item in x.get_ticklabels():
    item.set_rotation(90)
plt.show()

# adjust the subplot so the text doesn't run off the image
plt.subplots_adjust(bottom=0.3)
```
